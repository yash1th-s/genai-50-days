# Day 21: Image + Text Prompting

Same image. Different prompt. Completely different output.

Today we master prompting for vision.

## Setup

```python
from google import genai
from google.genai import types
import os
from dotenv import load_dotenv

load_dotenv(dotenv_path='../.env')
API_KEY = os.environ["GEMINI_API_KEY2"]
client = genai.Client(api_key=API_KEY)

IMAGE = 'image.png'

def ask(prompt):
    """Send image with prompt to Gemini."""
    with open(IMAGE, 'rb') as f:
        image_bytes = f.read()
    
    response = client.models.generate_content(
        model='gemini-2.5-flash-lite',
        contents=[
            types.Part.from_bytes(data=image_bytes, mime_type='image/png'),
            prompt
        ]
    )
    return response.text
```

## Length Control

Control how much output you get.

```python
print("ONE WORD:", ask("Describe this image in ONE word."))
print()
print("ONE SENTENCE:", ask("Describe this image in one sentence."))
print()
print("PARAGRAPH:", ask("Describe this image in detail."))
```

## Format Control

Get structured output.

```python
# JSON format
json_prompt = """Return a JSON object with:
- main_subject
- colors (list)
- mood

Return ONLY valid JSON."""

print(ask(json_prompt))
```

```python
# Bullet list
print(ask("Describe this image using bullet points: subject, colors, mood."))
```

## Task-Specific Prompts

```python
# Classification
print("CATEGORY:", ask("Classify: Nature, Urban, People, Food, Art, Other. Return only the category."))
print()

# Alt text
print("ALT TEXT:", ask("Write alt text for accessibility."))
print()

# Social media
print("HASHTAGS:", ask("Generate 5 hashtags for this image."))
```

## Role-Based Prompts

```python
# Art critic
print("ART CRITIC:")
print(ask("You are an art critic. Analyze composition and color in 2 sentences."))
print()

# Travel blogger
print("TRAVEL BLOGGER:")
print(ask("You are a travel blogger. Suggest a catchy blog title for this image."))
```

## Key Takeaways

1. **Length control** — One word, sentence, or paragraph
2. **Format control** — JSON, bullet list, any structure
3. **Task framing** — Classify, describe, generate hashtags
4. **Role-based** — Art critic, blogger, etc.
